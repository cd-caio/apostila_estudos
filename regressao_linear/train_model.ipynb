y = b0 + b1*x1 + b2*x2 + e
y^ = b0 + b1*x1 + b2*x2

y = valor real
y^ = valor da predição
b0, b1 e b2 = valores a serem estimados
x1 e x2 = variaveis independentes
e = erro intrínseco

Método dos minimos quadrados (MQO): Forma de calcular uma regressão linear fazendo com que a soma dos quadrados dos erros (SQE) seja a menor possível.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # biblioteca para visualização de dados
```

```python
df = pd.read_csv("train_data_final.csv")
df.columns
```

```python
sale_price = df["SalePrice"].to_numpy()
garage_area = df["LotArea"].to_numpy()
lot_area = df["GarageArea"].to_numpy()
```

```python
xdata = lot_area
ydata = sale_price
xbar  = np.mean(lot_area)
ybar  = np.mean(sale_price)


b0 = np.sum(xdata*(ydata-ybar))/np.sum(xdata*(xdata-xbar)) #coef. angular estimado
b1 = ybar - b0*xbar #coeficiente linear estimado

yest = b0*xdata + b1 # função estimada
```

```python
plt.figure()
plt.plot(xdata, yest, 'g', label = "fitted curve")
plt.scatter(xdata, ydata, label = "data")
plt.xlabel("Número de reclamações")
plt.ylabel("Pagamento total das reclamações [milhares de coroas suecas]")
plt.show()
```
